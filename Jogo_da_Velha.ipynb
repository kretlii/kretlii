```python
velha = [
      [0, 0, 0],
      [0, 0, 0],
      [0, 0, 0]
]

count = 0
start = 1

def inicio():

  print("***************************")
  print()
  print("      JOGO DA VELHA")
  print()
  print("*************************** \n")

  start = int(input(" INICIAR (1) // RESETAR (0) \n"))
  if start == 1:
    for i in range (5):

      jogo(1)
      check()
      if i == 4:
        print ("JOGO EMPATADO!!!")
        print ()
        break

      if start == 0:
        break

      jogo(-1)
      check()
      if start == 0:
        break

  else:
    print ("Volte sempre!!!")

## Foi definida uma função com variável "a", que se alterna entre 1 e -1, sendo
## 1 referente a vez do jogador "X" e -1 a do "O", em que o usuário da vez escolhe a posição desejada.

def jogo(a):

  if a == 1:
    print()
    print("### VEZ DE JOGADOR X ### \n")

    tela()
    print()

  elif a == -1:
    print()
    print("### VEZ DE JOGADOR O ### \n")

    tela()
    print()

  l = int(input("Digite linha desejada: "))
  c = int(input("Digite a coluna desejada: "))

  # Uma condicional afim que só preencha posições ja não ocupadas na matriz

  if velha[l - 1][c - 1] == 0:
    velha[l - 1][c - 1] = a
  else:
    print("Movimento inválido! Tente novamente. \n")
    tela()
    jogo(a)

## Função de checagem é a que realiza a somatória de todas as colunas, linhas e diagonais principais da matriz.
## Caso alguma das somas resultarem em 3 positivo o "X" ganha o jogo, caso contrário, é o "O" quem vence.

def check():

  # checando colunas
  for i in range (3):
    soma = velha[i][0] + velha[i][1] + velha[i][2]
    if soma == 3:
      print()
      print("Ganhador: JOGADOR 1")
      obg()
    elif soma == -3:
      print()
      print("Ganhador: JOGADOR 2")
      obg()
  soma = 0

  #checando linhas
  for i in range (3):
    soma = velha[0][i] + velha[1][i] + velha[2][i]
    if soma == 3:
      print()
      print("Ganhador: JOGADOR 1")
      obg()
    elif soma == -3:
      print()
      print("Ganhador: JOGADOR 2")
      obg()
  soma = 0

  #checando diagonal 1
  for i in range (3):
    soma = velha[0][0] + velha[1][1] + velha[2][2]
    if soma == 3:
      print()
      print("Ganhador: JOGADOR 1")
      obg()
    elif soma == -3:
      print()
      print("Ganhador: JOGADOR 2")
      obg()
  soma = 0

  #checando diagonal 2
  for i in range (3):
    soma = velha[0][2] + velha[1][1] + velha[2][0]
    if soma == 3:
      print()
      print("Ganhador: JOGADOR 1")
      obg()
    elif soma == -3:
      print()
      print("Ganhador: JOGADOR 2")
      obg()
  soma = 0

## A função tela é utilizada para mostrar ao usuário a matriz, substituindo os elementos numéricos pelo os do jogo.

def tela():

  print()
  for i in range (3):
    for j in range(3):
      if velha[i][j] == 0:
        print("- ", end = '')
      elif velha[i][j] == 1:
        print("X ", end = '')
      elif velha[i][j] == -1:
        print("O ", end = '')

    print()

## A função obrigado reseta a matriz para seu valor inicial e desliga o loop start do início
## após isso, mostra novamente a tela inicial do jogo

def obg():
  tela()

  for i in range (3):
    for j in range (3):
      velha [i][j] = 0

  print()
  print("**************************")
  print()
  print("Obrigado pela preferência! \n")

  start = 0

  inicio()

inicio()


```
